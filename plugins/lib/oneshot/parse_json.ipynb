```python
import pandas as pd
import json
```

```python
with open('zozotown_category.json') as f:
    data = json.load(f)
```

```python
category_dict = data['categorydata']
top_category = list(category_dict.keys())
```

```python
category_df = pd.DataFrame(
    columns=[
    'category_id', 
    'category_name', 
    'category_path', 
    ])
```

```python
child_category_df = pd.DataFrame(
    columns=[
    'category_id',
    'category_name',
    'category_path',
    'parent_category_id',
    ])
```

```python
for top in top_category:
    category_df = category_df.append(
        {
        'category_id': top,
        'category_name': category_dict[top]['name'],
        'category_path': category_dict[top]['path'],
        },
        ignore_index=True
    )
    for child in category_dict[top]["type"]:
        child_category_df = child_category_df.append(
            {
            'category_id': child['id'],
            'category_name': child['name'],
            'category_path': child['path'],
            'parent_category_id': top,
            },
            ignore_index=True
        )
```

```python
category_df.to_csv('zozotown_category.csv', index=False)
child_category_df.to_csv('zozotown_child_category.csv', index=False)
```
